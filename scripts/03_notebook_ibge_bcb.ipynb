 Atividade IBGE + Banco Central

IBGE: API de regiões do Brasil  
BCB: Série 433 (índice econômico)

```python
import requests
import pandas as pd
```

```python
url = "https://servicodados.ibge.gov.br/api/v1/localidades/regioes"

response = requests.get(url)
data = response.json()

df_ibge = pd.json_normalize(data)

df_ibge.head()
```

```python
url = "https://api.bcb.gov.br/dados/serie/bcdata.sgs.433/dados?formato=json"

response = requests.get(url)
dados = response.json()

df = pd.DataFrame(dados)

df.head()
```

Tratamento: conversão de tipos e limpeza de dados

```python
df["valor"] = pd.to_numeric(df["valor"], errors="coerce")
df["data"] = pd.to_datetime(df["data"], dayfirst=True)
```

```python
df.info()
```

```python
df.to_csv("bcb_tabela.csv", index=False)
df.to_csv("dados_tratados.csv", index=False)
```

Scraping: técnica para extrair dados de sites sem API.

Riscos: violação de termos de uso, direitos autorais e LGPD.
Mitigação: usar dados públicos e respeitar regras do site.

API vs scraping: API é estruturada, estável e oficial.
Scraping é mais frágil pois depende do HTML.
A melhor abordagem é usar API quando disponível.
